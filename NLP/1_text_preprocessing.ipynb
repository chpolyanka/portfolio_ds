# 3. Выполним предварительную обработку текста

```python
pip install pymorphy2
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
import pymorphy2
from nltk.tokenize import word_tokenize
import re


nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```python
# Инициализация морфологического анализатора
morph = pymorphy2.MorphAnalyzer()
def preprocess_text(text):
    # Удаление лишних символов (оставляем только буквы и пробелы)
    text = re.sub(r'[^а-яА-ЯёЁ\s]', '', text)
    # Приведение к нижнему регистру
    text = text.lower()
    # Токенизация
    tokens = word_tokenize(text)
    # Удаление стоп-слов
    stop_words_set = set(stopwords.words('russian'))
    tokens = [token for token in tokens if token not in stop_words_set]
    # Лемматизация каждого слова
    lemmatized_tokens = [morph.parse(token)[0].normal_form for token in tokens]
    # Возвращаем обработанный текст
    return ' '.join(lemmatized_tokens)
```

```python
# Применение функции предварительной обработки к каждой записи в столбце 'review'
data['processed_review'] = text.apply(preprocess_text)

# Вывод результатов
data['processed_review']
```

# 4. Топ-10 наиболее часто используемых слов в отзывах

```python
from collections import Counter

# Объединяем все обработанные отзывы в один текстовый блок
all_words = ' '.join(data['processed_review']).split()

# Подсчет частоты слов
word_counts = Counter(all_words)

# Находим топ-10 наиболее часто используемых слов
most_common_words = word_counts.most_common(10)

most_common_words
```

# 5. Положительные и отрицательные отзывы

```python
# Преобразование типа данных столбца 'score' в числовой
data['score'] = pd.to_numeric(data['score'], errors='coerce')
```

```python
# Отбор отрицательных отзывов (с оценками 1 и 2)
negative_reviews = data[data['score'] <= 2]

# Отбор положительных отзывов (с оценками 4 и 5)
positive_reviews = data[data['score'] >= 4]
```

# 6. Топ-10 наиболее часто используемых прилагательных, используемых со словом банк

```python
def find_adj_bank(reviews, morph):
    adjectives = []
    for review in reviews:
        words = review.split()
        for i, word in enumerate(words):
            if word == "банк":
                #прилагательные перед и после слова "банк"
                for j in range(max(0, i-3), min(i+4, len(words))):
                    parsed_word = morph.parse(words[j])[0]
                    if parsed_word.tag.POS in ['ADJF', 'ADJS']:
                        adjectives.append(parsed_word.normal_form)
    return adjectives

morph = pymorphy2.MorphAnalyzer()

# Извлекаем списки прилагательных для положительных и отрицательных отзывов
positive_adjectives = find_adj_bank(positive_reviews['processed_review'], morph)
negative_adjectives = find_adj_bank(negative_reviews['processed_review'], morph)

top_positive_adjectives = Counter(positive_adjectives).most_common(10)
top_negative_adjectives = Counter(negative_adjectives).most_common(10)

print("Топ-10 положительных прилагательных:", top_positive_adjectives)
print("Топ-10 отрицательных прилагательных:", top_negative_adjectives)
```

Положительные отзывы чаще всего содержат прилагательное "хороший", что указывает на удовлетворенность клиентов. Также встречаются слова "нужный" и "огромный", что может говорить о значимости предоставляемых банком услуг.

В отрицательных отзывах доминируют слова, указывающие на персональное или субъективное восприятие, такие как "мой", "ваш", "свой".
